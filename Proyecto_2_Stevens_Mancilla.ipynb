```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
Link = 'https://github.com/WhosLupe/THC_proyect/blob/main/ebola_2014_2016_clean.csv'
code = '?raw=true'
url = Link+code
df = pd.read_csv(url,sep=',',index_col=1)
```

```python
PIvsD = 100*SL['No. of confirmed deaths']/SL['No. of confirmed cases']
PIvsD
```

```python
SL['PIvsD']=PIvsD
```

```python
recuperados = ((100-PIvsD)/100)*SL['No. of confirmed cases']
SL['recuperados'] = recuperados
```

```python
SL['PR'] = (100-PIvsD)
```

```python
PIvsD_mean = SL['PIvsD'].mean()
PR_mean = SL['PR'].mean()
porcentajes = [PIvsD_mean,PR_mean]
etiquetas = ['Descesos','Recuperaciones']
```

```python
plt.pie(porcentajes,labels=etiquetas,autopct='%0.1f %%', colors=['gray','pink'])#Colores del gráfico
plt.axis("equal")
plt.title('Porcentaje entre recuperados vs decesos en Italia')
plt.savefig("Pastel_PIvsD_PR", dpi = 300,format = 'png')
plt.show
```
